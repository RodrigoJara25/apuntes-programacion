# Uso de ELSE con el bloque TRY-EXCEPT

El bloque ELSE se ejecutará cuando no ocurrió excepcion alguna

```python
# Forcemos una excepcion

try:
  x = 3.14159/0.0
except:
  print("Usario : Definiste la variable usandoo una operacion inadecuada")
  x = float(input("Ingresa el numerador : "))
  y = float(input("Ingresa el denominador  "))
else:
  print("Usario : No a ocurrido ninguna excepcion")
```

```python
# Forcemos una excepcion

try:
  x = 3.14159/2.0
except:
  print("Usario : Definiste la variable usandoo una operacion inadecuada")
  x = float(input("Ingresa el numerador : "))
  y = float(input("Ingresa el denominador  "))
else:
  print("Usario : No a ocurrido ninguna excepcion")
```

# La palabra reservada FINALLY

Asociada a un bloque try-except-else tambien le podemos asociar un bloque finally. Este bloque finally se ejecuta siempre (haya o no ocurrido excepcion alguna)

```python
# Veamos la excepcion que se obtiene de intentar cargar un archivo de datos
# en memoria sin que este este disponible
with open("data1.txt") as file:
  data_obtenida = file_read()
```

```python
# Intentemos abrir (carga en memoria) un archivo de datos de nuestro sistema
# de archivos

try :
  with open("data1.txt") as file:
    data_obtenida = file_read()
except FileNotFoundError:
  print("Se lanzo una excepcion : No se puede cargar en memoria la base de datos")
except OSError:
  print("Esto es un error del sistema operativo")
```

```python
# Bloque try-except-else-finally

try :
  with open("data1.txt") as file:
    data_obtenida = file_read()
except FileNotFoundError:
  print("Se lanzo una excepcion : No se puede cargar en memoria la base de datos")
except OSError:
  print("Esto es un error del sistema operativo")
else:
  print("No ha ocurrido excepcion alguna")
finally:
  import os
  os.system("wget https://raw.githubusercontent.com/robintux/Datasets4StackOverFlowQuestions/master/ACS.csv")
```
