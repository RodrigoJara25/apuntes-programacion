```python
# Implementacion de clase Animal
class Animal():
  def __init__(self, edad = 0, nombre = "Godzilla"):
    """
    Constructor de la clase Animal

    :param edad : Edad del animal : Por defecto tiene el valor de 0
    :param nombre : Nombre del animal : Por defecto tiene el valor de Godzilla
    :return : Una instancia de la clase Animal
    """

    # Todos los atributos del constructor sean publicos
    self.edad = edad
    self.nombre = nombre
    self.sexo = "F"
    self.vivo = True
    # Animal.numAnimals = Animal.numAnimals + 1   # contador de veces que se ha instanciado esta clase

  def saluda():
    print("Hola Mundo")

  def showmethename(self):
    print("Bienvenido " + self.nombre)

  def showmetheage(self):
    print("Edad " + str(self.edad))
```

```python
#
anm1 = Animal(edad = 12, nombre = "Firulais")
```
