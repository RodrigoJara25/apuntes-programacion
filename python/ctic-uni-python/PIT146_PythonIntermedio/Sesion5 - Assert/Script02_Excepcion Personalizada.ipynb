```python
# Para crear/definir/implementar nuestra propia excepcion debemos considerar la clase Exception
# Esta clase Exception, como en otras vistas, va a ser la clase padre de nuestra propia excepcion

# Definir una subclase : MiExcepcionPersonalizada
# Clase padre : Exception

class MiExcepcionPersonalizada(Exception):
    pass
```

```python
# Lancemos/ejecutemos/instanciemos mi subclase personalziada
raise MiExcepcionPersonalizada()
```

```python
# Documentacion MiExcepcionPersonalizada
help(MiExcepcionPersonalizada)
```

```python
# Lista de metodos que se puede aplicar a un objeto que es instancia de la clase MiExcepcionPersonalizada
x1 = MiExcepcionPersonalizada()
dir(x1)
```

```python
# Tambien podemos pasar argumentos de entrada al lanzar nuestra subclase
# Ejemplo : Pasemos dos argumentos al lanzar nuestra excepcion. Para ello necesitamos
# modificar la funcion __init__() de la siguiente manera

# Modifiquemos el metodo __init()__ de nuestra excepcion personalizada
class MiExcepcionPersonalizada(Exception):
  def __init__(self, parametro1, parametro2):
    self.parametro1 = parametro1
    self.parametro2 = parametro2
```

```python
# __init()__ se ejecuta cuando instancio un objeto de esa clase
# siempre el primer argumento es "self"
```

```python
#
x2 = MiExcepcionPersonalizada("Python", "Intermedio")
```
