# La estructura de repetición while

```python
contador = 0
while contador < 10:
    print('phyton' + "\tValor actual de contador " + str(contador))
    contador = contador + 2
```

```python
# Considere el escenario donde el valor de actualizacion es un aleatorio

import random as rnd

numveces = 0
numIter = 0
while numveces < 20:
  numIter = numIter + 1
  print("Phyton ", "Numero de veces " + str(numveces))
  numveces = numveces + (rnd.randint(0,20))**(1/20)
print("Numero de iteraciones ", numIter)
```

```python
# Pidamos al usuario un número diferente de 0
# Si ingresa 0 debe pedir nuevamente el numero
#
# UPDATE al requerimiento: Solo puede equivocar 3 veces consecutivas

num = int(input("Ingrese un numero distinto a 0: "))
NumVezErro = 0
while num == 0:
  num = int(input("Ingrese un numero distinto a 0: "))
  NumVezErro = NumVezErro + 1
  if NumVezErro == 2:
    print("Se te acabaron los intentos")
    break
```

```python
# Ingresar un conjunto no determinado de notas y calcular el promedio
# El valor 999 indica el fin del ingreso de notas

nota = float(input("Ingrese una nota: "))
if nota != 999:
  suma = 0
  contador = 0
  while nota != 999:
    contador = contador + 1
    suma = suma + nota
    nota = float(input("Ingrese una nota: "))
  promedio = suma/contador
  print("Promedio: ", promedio)
else:
  print("No se ingresaron notas")
```

# La estructura MATCH

```python
#match parametro:
#  case valor1:
#    sentencia1
#    sentencia2
#  case valor2:
#    sentencia3
#    sentencia4
#  case _:
#    sentencia_por_defecto
```

```python
# Pidamos un numero al usuario
num = int(input("Ingresa un numero: "))

match num:
  case 1:
    print("Ingresaste el numero 1")
  case 2:
    print("Ingresaste el numero 2")
  case 3:
    print("Ingresaste el numero 3")
  case _:
    print("No es 1 , 2 o 3")
```

```python
#
nombre = input("Ingresa tu nombre: ")

match nombre:
  case "Abraham" | "abraham":
    print("Usuario existente")
  case "Elon" | "elon":
    print("Usuario administrador")
  case "Anderson" | "anderson":
    print("Pogramador Senior")
  case _:
    print("Usuario Inexistente")
```

```python
#
lado_a = float(input("Ingresa el primer lado del triángulo "))

match lado_a:
  case lado_a if lado_a > 0:
    print("positivo")
  case lado_a if lado_a < 0:
    print("negativo")
  case _:
    print("cero")
```
