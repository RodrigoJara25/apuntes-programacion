```python
def biseccion(f,a,b,N):
  """
  Implementación de la estretgia explicada en clase : Metodo de la biseccion

  Parámetros de entrada
  ---------------------

  f : funcion (regla de correspondencia)
  a,b : numeros reales (representan el [a,b])
  N : Número entero (representa el numero de iteraciones)

  Salida
  ------
    m_n : float (aproximacion de la raiz f en [a,b])

  """

  # Verificar que el producto de imagenes de a y b cumplen la condicion del teorema
  if f(a)*f(b) >= 0:
    print("El metodo de la biseccion falla")
    print("El intervalo no es el adecuado")
    return None

  # Consideremos que se cumplen las condiciones del teorema
  a_n = a
  b_n = b

  for n in range (1,N+1):
    # Calculamos el punto medio
    m_n = (a_n + b_n)/2
    # Calculemos la imagen mediante f de ese punto medio
    f_m_n = f(m_n)
    # Seleccionar el subintervalo izquierdo
    if f(a_n)*f_m_n < 0:
      a_n = a_n
      b_n = m_n
    # Seleccionar el subintervalo derecho
    elif f(b_n)*f_m_n < 0:
      a_n = m_n
      b_n = b_n
    # El punto medio sea solucion exacta
    elif f_m_n == 0:
      print("Solucion exacta")
      return m_n
    else:
      print("El metodo de la biseccion falla")
      return None
  return (a_n + b_n) / 2
```

```python
import math as m

def f(x):
  return x**(m.sin(x/3.0)) - (1/3.0)

x = biseccion(f,0,1,40)
x

# respuesta matemática = 1.0001

# [5,15] = 10.860801790383903
# [10,15] = 10.860801790383903

# [infinito, 15] = 10.860801790383903
# [15, infitivo] = 17.671963658929144
```
