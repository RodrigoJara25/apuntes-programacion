Carga del módulo matemático

```python
# De esta forma, se carga cualquier módulo en phyton

import math as m
```

```python
# Documentación del módulo math (alias m)
# En lenguajes compilados se les llama librerias, pero en lenguajes interpretados se llama modulo

help(m)

# "help" sirve para mostrar documentación
```

```python
# Lista de componentes del modulo matematico
dir(m)
# Me interesan las que no tienen "_"
# CTRL + SLASH
```

```python
# Documentacion de la funcion radians
help(m.radians)
```

```python
# Documentacion de la funcion hypot
help(m.hypot)
```
