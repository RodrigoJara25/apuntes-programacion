```python
# Implementa una funcion que genere un numero determinado de aleatorios
# en una lista y calcule su promedio geométrico

import random as rnd

def MakeList(num):
  var1 = []
  for i in range(num):
    var1.append(rnd.randint(0,20))
  return var1

def CalculoMG(lista):
  prod = 1
  for elemento in lista:
    prod = prod * elemento
  return (prod)**(1/len(lista))
```

```python
CalculoMG(MakeList(10))
```

# Método de la Biseccion

image.png

```python
def biseccion(f,a,b,N):
  """
  Implementación de la estretgia explicada en clase : Metodo de la biseccion

  Parámetros de entrada
  ---------------------

  f : funcion (regla de correspondencia)
  a,b : numeros reales (representan el [a,b])
  N : Número entero (representa el numero de iteraciones)

  Salida
  ------
    m_n : float (aproximacion de la raiz f en [a,b])

  """

  # Verificar que el producto de imagenes de a y b cumplen la condicion del teorema
  if f(a)*f(b) >= 0:
    print("El metodo de la biseccion falla")
    print("El intervalo no es el adecuado")
    return None

  # Consideremos que se cumplen las condiciones del teorema
  a_n = a
  b_n = b

  for n in range (1,N+1):
    # Calculamos el punto medio
    m_n = (a_n + b_n)/2
    # Calculemos la imagen mediante f de ese punto medio
    f_m_n = f(m_n)
    # Seleccionar el subintervalo izquierdo
    if f(a_n)*f_m_n < 0:
      a_n = a_n
      b_n = m_n
    # Seleccionar el subintervalo derecho
    elif f(b_n)*f_m_n < 0:
      a_n = m_n
      b_n = b_n
    # El punto medio sea solucion exacta
    elif f_m_n == 0:
      print("Solucion exacta")
      return m_n
    else:
      print("El metodo de la biseccion falla")
      return None
  return (a_n + b_n) / 2

```

```python
# Ecuacion cuadratica
def f1(x):
  return x**2 -x -1

# Raiz aproximada
c_sombrero = biseccion(f1,1,2,50)

# Calculemos la aproximacion de una raiz
print("""
Biseccion : %.3f
f(c) : %.10f
""" %(c_sombrero, f1(c_sombrero)))

```

```python
# Ecuacion cúbica
def f2(x):
  return x**3 - x**2 +2

# Raiz aproximada
c_sombrero = biseccion(f2,-200,300,50)

# Calculemos la aproximacion de una raiz
print("""
Biseccion : %.3f
f(c) : %.10f
""" %(c_sombrero, f2(c_sombrero)))
```

```python
# Aproximacion a raices cuadradas (raiz de 3)
def f3(x):
  return x*x - 3

# Raiz aproximada
c_sombrero = biseccion(f3,1,2,50)

# Calculemos la aproximacion de una raiz
print("""
Biseccion : %.3f
f(c) : %.10f
""" %(c_sombrero, f3(c_sombrero)))
```

```python
# Consideremos una regla de correspondencia para al funcion un poco mas algebraica

import math as m
def f4(x):
  return m.exp(x)*(3.2*m.sin(x) - 0.5*m.cos(x))

# Aproximacion a la raiz
x1 = biseccion (f4, 0,2,50)
x2 = biseccion (f4, 2,4,50)

print("""
En [0,2] : c = %f
En [2,4] : c = %f
""" %(x1,x2))

```
