```python
# Documentacion de la funcion inv
# Calcula la inversa multiplicativa de cada elemento de una matriz

help(np.linalg.inv)
```

```python
# Calculamos la inversa de la matriz matNotas
matInv = np.linalg.inv(matNotas)
matInv
```

```python
# Cumple la definicion operativa de la matriz inversa
np.dot(matNotas, matInv)

# Me da la matriz identidad
```

```python
# Calculemos la inversa de mat1
Inversa_mat1 = np.linalg.inv(mat1)
Inversa_mat1
```

```python
# Verifiquemos (dot : producto escalar de dos matrices)
np.dot(mat1, Inversa_mat1)
```
