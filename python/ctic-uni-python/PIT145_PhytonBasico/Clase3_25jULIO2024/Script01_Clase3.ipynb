## Acceso a los elementos de un dato de tipo str

```python
# Un str es un objeto con naturaleza vectorial: El acceso
# es por medio de indices.

# En phyton, los indices inician en cero.
```

```python
# La primera componente (primer caracter) de un str es el indice cero
cad1[0]
```

```python
# La segunda componente es el indice 1
cad1[1]
```

```python
# Conlusion 1: los indices en phyton (leidos de izquierda a derecha) inician
# en 0 y van aumentando de 1 en 1
```

```python
# Notar que tambien podemos acceder a los elementos de un str de
# derecha a izquierda, en cuyo caso el conjunto de indices incia
# en -1 y va disminuyendo de -1 en -1
```

```python
# Última componente de cad3
cad3[-1]
```

```python
# Penultima componente de cad3
cad3[-2]
```

```python
# Acceso a porciones (slicings) de str: usamos la siguiente notacion
```

```python
# Ultimos 7 caracteres de cad2
cad2[-7:]

# Reconocer que los ":" al final significa que llega al final del str
# Al indice se le suma +1. Si quieres 10 elementos, el indice es 9, pero
# por sintaxis se le suma 1.
```

```python
# Primeros 50 caracteres de cad3
cad3[:49+1]
```

```python
# Primeros 10 caracteres de cad2
cad2[:9+1]
```

```python
# Obtengamos los caracteres del indice 8 hasta el indice 61 de cad3
cad3[8:61+1]
```

## Operadores lógicos para "bool"
- AND
- OR
- NOT

```python
# and: y logico
# or: o logico
# not: negacion

# and y or son operaciones binarias (aplican sobre dos bool)
# and: bool x bool -> bool
# or: bool x bool -> bool

# not (operacion booleana de negacion) es una operacion unaria
# not: bool -> bool

b10 =  b1 or b2
b11 = b1 and b2
b12 = not b11
```
