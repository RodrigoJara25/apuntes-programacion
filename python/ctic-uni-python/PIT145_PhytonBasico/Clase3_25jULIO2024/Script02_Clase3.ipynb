# Rutina de entrada
- Funcion input

```python
# Pedir informacion al usuario (siempre se almacena como string por default)

# Ingresar nombre
nombre = input("Ingresa tu nombre: ")
```

```python
# tipo de dato
type(nombre)
```

```python
# Ingresar edad
edad = input("Ingresa tu edad: ")
```

```python
# tipo de dato
type(edad)
```

```python
# Phyton es un lenguaje de programacion de tipado dinamico
edad = int(edad)
edad
```

```python
# Ingresar estatura en metros (punto flotante)
estatura = float(input("Ingresa tu estatura en metros: "))
estatura
```

```python
peso = float(input("Ingresa tu peso: "))
peso
```

```python
# Calculo del indice de masa corporal
IMC = peso/(estatura**2)
IMC
```

# Rutina de salida
- Funcion print

```python
# Primera forma de uso
print(nombre)
print(edad)
print(IMC)
```

```python
# Segunda forma de uso
print("Nombre ingresado ", nombre)
print("edad: ", edad, "Estatura: ", estatura)
```

```python
# Tercera forma de uso:
# \n : salto de linea
# \t : tabular

print("Nombre ingresado\n", nombre)
print("edad : ", edad, "\nEstatura : ", estatura)
```

```python
# Cuarta y quinta forma de uso de la funcion print :
# docstring con % de C
print("""
Nombre ingresado : %s
Edad : %i
Estatura : %.2f
IMC : %.2f
""" %(nombre, edad, estatura, IMC))

# %s : str
# %i : int
# %f : float
```
