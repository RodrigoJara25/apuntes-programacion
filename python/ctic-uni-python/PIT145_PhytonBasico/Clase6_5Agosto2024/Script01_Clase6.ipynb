```python
# Primero funcion

def calculoArea(lado1, lado2, lado3):
  """
  Funcion que calcula el area de un triangulo
  """
  s = (lado1+lado2+lado3)/2
  area = (s*(s-lado1)*(s-lado2)*(s-lado3))**0.5
  return area
```

```python
# Prueba 1
calculoArea(12,19,21)
```

```python
# Funcion que calcula el area y el perimetro de un rectangulo
def CalculoAreaPerimetro(base, altura):
  perimetro = 2*base + 2*altura
  area = base*altura
  return perimetro, area
```

```python
# Prueba de CalculoAreaPerimetro
b = 12.8
h = 9.81
p1, a1 = CalculoAreaPerimetro(b,h)
print("""
El perimetro es: %.3f
El area es: %.3f
""" %(p1,a1))
```

```python
# Dada una acdena de caracteres y un caracter
# Calcular el numero de veces que aparece el caracter en la
# acdena de caracteres

def CuentaCaracteres(cadena, caracter):
  Numveces = cadena.count(caracter)
  return Numveces
```

```python
cad = "Esto es una prueba para la funcion definida en la sexta clase"
a = CuentaCaracteres(cad, "a")
e = CuentaCaracteres(cad, "e")
i = CuentaCaracteres(cad, "i")
o = CuentaCaracteres(cad, "o")
u = CuentaCaracteres(cad, "u")

print("""
Numero de veces
Vocal a: %d
Vocal e: %d
Vocal i: %d
Vocal o: %d
Vocal u: %d
""" %(a,e,i,o,u))
```

```python
# Dada una cadena de caracteres. Implementar una funcion que genere
# una lista de palabras aleatorias.

molicie = """
Lorem Ipsum es simplemente el texto de relleno de las imprentas y
archivos de texto. Lorem Ipsum ha sido el texto de relleno estándar
de las industrias desde el año 1500, cuando un impresor (N. del T.
persona que se dedica a la imprenta) desconocido usó una galería de
textos y los mezcló de tal manera que logró hacer un libro de textos
especimen. No sólo sobrevivió 500 años, sino que tambien ingresó como
texto de relleno en documentos electrónicos, quedando esencialmente
igual al original. Fue popularizado en los 60s con la creación de las
hojas "Letraset", las cuales contenian pasajes de Lorem Ipsum, y más
recientemente con software de autoedición, como por ejemplo Aldus
PageMaker, el cual incluye versiones de Lorem Ipsum.
"""
```

```python
# Cargamos cada palabra de la cadena como un elemento de la lista
molicie.split()
```

```python
# Funcion que geenra una lista de palabras aleatorias a aprtir de una
# cadena de caracteres

def GeneraListaPalabras(cadena, num):
  """
  cad: una cadena de caracteres
  num: numero de elementos random de la lista a generar
  """
  numPalabras = len(cad.split())
  listaP = []
  for c in range(num):
    indice = rnd.randint(0,numPalabras-1)
    listaP.append(cad.split()[indice])
  return listaP
```

```python
# Lista de 12 palabras aleatorias
GeneraListaPalabras(molicie, 12)
```
